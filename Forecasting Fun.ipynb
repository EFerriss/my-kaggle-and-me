## Predicting Sales

Goal: predict total sales for **every product** and **every store** in the next month. The list of shops and products slightly changes every month.  
Evaluation metric: [RMSE](https://en.wikipedia.org/wiki/Root-mean-square_deviation)

```python
def predict_sales(shop_id, item_id):
    train_shop = train[train.shop_id == shop_id]
    train_item_shop = train_shop[train_shop.item_id == item_id]
    print(len(train))
    print(len(train_shop))
    print(len(train_item_shop))
    
    return train_item_shop
    
df = predict_sales(12, 11373)
```
